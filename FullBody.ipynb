```python
import cv2
from google.colab import drive
drive.mount ('/content/drive')
from google.colab.patches import cv2_imshow
```

```python
human = cv2.imread('/content/drive/MyDrive/ComputerVision/images/people3.jpg')
human_gray = cv2.cvtColor(human, cv2.COLOR_BGR2GRAY)
human_detector = cv2.CascadeClassifier('/content/drive/MyDrive/ComputerVision/Cascade/fullbody.xml')
human_detection = human_detector.detectMultiScale(human_gray, scaleFactor = 1.01,
                                                  minSize=(45,100))
for (x, y, w, h) in human_detection:
  cv2.rectangle(human, (x, y), (x + w, y + h), (0, 225, 225), 2)
  print(w,h)

cv2_imshow(human)
```
