```python
import cv2
from google.colab import drive
drive.mount ('/content/drive')
from google.colab.patches import cv2_imshow
```

```python
car = cv2.imread('/content/drive/MyDrive/ComputerVision/images/car.jpg')
car_gray = cv2.cvtColor(car, cv2.COLOR_BGR2GRAY)
car_detector = cv2.CascadeClassifier('/content/drive/MyDrive/ComputerVision/Cascade/cars.xml')
detection = car_detector.detectMultiScale(car_gray, scaleFactor = 1.008, minNeighbors=10,
                                          minSize = (31, 31))
for (x, y, w, h) in detection:
  cv2.rectangle(car, (x, y), (x + w, y + h), (0, 0, 255), 3)
  print(w, h)
cv2_imshow(car)
```
